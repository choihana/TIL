```python
from zipfile import ZipFile
z = ZipFile('sentiment.zip')
data = z.open('sentiment labelled sentences/imdb_labelled.txt')
```

```python
import pandas as pd
df = pd.read_csv(data, sep='\t', header = None)
```

## 감성 분석 2가지 방법

1. 감성 단어를 알려주기: good -> 긍정, bad -> 부정
2. 기계학습


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=500, stop_words='english')
tdm = cv.fit_transform(df[0])
```

```python
words = cv.get_feature_names()
words
```

## 긍정 점수

```python
# 워드 사전에서 good 단어의 인덱스
# 181열 
words.index('good')
```

```python
# 워드 사전에서 fun 단어의 인덱스
# 164열
words.index('fun')
```

```python
# 긍정 단어 인덱스
pos_index = [181,160]
```

```python
# 긍정 단어 빈도수 count 를 긍정 score라고 한다
pos_score = tdm[:,pos_index].sum(axis =1)
```

```python
pos_score > 1
```

## 부정 점수

```python
# 워드 사전에서 bad 단어의 인덱스
# 31열
words.index('bad')
```

```python
# 워드 사전에서 boring 단어의 인덱스
# 46열
words.index('boring')
```

```python
neg_idx = [31,46]
```

```python
neg_score = tdm[:,neg_idx].sum(axis=1)
```

## 예측

```python
import numpy as np
```

```python
y_pred = np.where(pos_score > neg_score, 1, 0)
```

## 정확도 평가

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(df[1],y_pred)
```

```python
df[1].mean()
```

```python
y_pred.mean()
```

## 단어추가해서 성능올리기

```python
## 전처리만 해도 성능 올라갈 것 
```

```python
# 긍정 단어 인덱스
pos_idx = [words.index('good')
             ,words.index('fun')
            ,words.index('best')
            ,words.index('love')
          ,words.index('adorable')
          ,words.index('amazing')
          ,words.index('beautiful')
          ,words.index('great')
          ,words.index('wonderful')
          ,words.index('funny')
          ,words.index('excellent')]
```

```python
# 부정 단어 인덱스 
neg_idx = [words.index('bad')
             ,words.index('boring')
            ,words.index('lacks')
            ,words.index('awful')
          ,words.index('annoying')]
```

```python
pos_score = tdm[:,pos_index].sum(axis =1)
neg_score = tdm[:,neg_idx].sum(axis=1)
```

```python
y_pred = np.where(pos_score > neg_score, 1, 0)
```

```python
accuracy_score(df[1],y_pred)
```

단어사전보다 기계학습이 효율적이다!
그리고 전처리 과정을 넣어주면 더 성능 올라갈것
