```python
import pandas as pd
from wordcloud import WordCloud
```

```python
df = pd.read_csv('word_cout.csv')
```

`wordcloud`의 주요옵션  

- font_path: 글꼴의 경로
- max_words: 워드클라우드를 그릴 단어의 개수
- background_color: 배경색 설정
- width : 가로크기(픽셀 단위)
- height: 세로크기(픽셀 단위)

```python
wc = WordCloud(background_color='white', max_words=100, width=400, height=300)
```

```python
count_dic=df.set_index('단어')['빈도'].to_dict()
count_dic
```

```python
cloud = wc.fit_words(count_dic)
```

```python
cloud.to_image()
```

```python
# 저장
cloud.to_file('cloud.png')
```
