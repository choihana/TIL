## 당뇨병 데이터 셋 사용하기

출처: <a href = 'https://www.kaggle.com/uciml/pima-indians-diabetes-database'> 캐글 데이터 셋 </a>

문제: 독립변수를 사용하여 당뇨병 여부 파악하기 (분류 문제)

전처리나, 파라미터 조정 없이 train, test set으로 분류하여 decision tree로 예측해봅니다.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../data/diabetes.csv')
df.head()
```

### 데이터 구성

- Pregnancies: 임신 횟수
- Glucose: 2시간 동안의 경구 포도당 내성 검사에서 혈장 포도당 농도
- BloodPressure: 이완기 혈압 (mm Hg)
- Skin Thickness: 삼두근 피부 주름 두께 (mm), 체지방을 추정하는데 사용되는 값
- Insulin: 2시간 혈청 인슐린 (mu U /ml)
- BMI: 체질량 지수(kg / (m)^2)
- DiabetesPedigreeFunction: 당뇨병 혈통 기능
- Age: 나이
- Outcome: (예측해야할 종속 변수) 768개 중에 268개의 결과 클래스 변수(0또는 1)는 1이고 나머지는 0 입니다.

### train, test 데이터 셋 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[ : , :-1] #독립변수
y = df.iloc[ : ,-1: ] #종속변수
```

```python
X.head()
```

```python
y.head()
```

```python
# test size = 0.2, train size = 0.8
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 1111)
```

```python
print('train shape:',X_train.shape)
print('test shape:',X_test.shape)
```

### Decision Tree 가져오기

```python
from sklearn.tree import DecisionTreeClassifier

# 의사결정 나무 모델 만들기
model = DecisionTreeClassifier()
```

### 학습 시키기: fit

```python
model.fit(X_train, y_train)
```

### 예측하기 : predict

```python
pred = model.predict(X_test)
pred
```

### 트리 알고리즘 분석하기

- 시각화 해보기

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (20,20))
tree = plot_tree(model, feature_names = X.columns,
                filled = True,
                fontsize = 10)
```

### 피쳐 중요도 확인하기

```python
# 피쳐의 중요도
model.feature_importances_
```

```python
# Glucose와 BMI가 가장 중요한 피쳐임을 알 수 있음
sns.barplot(x=model.feature_importances_,y=X.columns)
```

### 모델 정확도 (Accuracy) 측정하기

- 전체 데이터 중 정답을 맞춘 확률 (True -> True, False -> False)

```python
pred
```

```python
from sklearn.metrics import accuracy_score
# 정확도
accuracy_score(y_test,pred)
```
