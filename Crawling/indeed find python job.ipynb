```python
# title, company 정보 크롤링
def extract_title_company(result):
    title=result.find('h2').find('a')['title']
    company=result.find('span',{'class':'company'}).text.strip()
    return {'title':title,'company':company}
```

```python
# 모든 페이지에서 추출
def extract_jobs(last_pages):
    jobs = []
    
    # 1~마지막 페이지 돌기
    for page in range(last_pages):
        #url변경으로 돌기
        result=requests.get(f'{url}&start={page*limit}')
        soup = bs(result.text, 'html.parser')
        results = soup.find_all('div',{'class','jobsearch-SerpJobCard'})
        # 각 div에서 title, company 가져와서 리스트에 추가하기
        for result in results:
            job = extract_title_company(result)
            print(job)
            jobs.append(job)
    return jobs
```

```python
last=extract_pages(url)
jobs=extract_jobs(last)
```
