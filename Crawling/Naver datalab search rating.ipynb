## selenium

동적 크롤링을 위한 라이브러리

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('./chromedriver')
#페이지가 모두 로드될 때까지 3초 기다려주는 코드
driver.implicitly_wait(3)
#셀레니움을 통해 실제 크롬창에서 접소갛ㄴ 것과 동일하게 해당 url 내용을 가져옴
driver.get(url)
#소스를 읽어 Beautifulsoup을 이용해 파싱
html = driver.page_source
soup = bs(html,'html.parser')

# 그다음은 이전에 requests와 beautifulsoup 을 이용했던 코드와 동일하게 처리

rank_list = soup.select('.ranking_list .item_title')
rank_result_list = []

for (index, rank) in enumerate(rank_list):
    rank_result_list.append({'rank':index+1,'value':rank.text})
print(rank_result_list)

driver.close()
```

### covid 19 동적크롤링

```python
driver = webdriver.Chrome('./chromedriver')
driver.implicitly_wait(3)
url = 'https://news.google.com/covid19/map?hl=ko&mid=%2Fm%2F02j71&gl=KR&ceid=KR%3Ako'
driver.get(url)
################################################################

#소스를 읽어 Beautifulsoup을 이용해 파싱
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')

# 스크롤을 하기위한 영역 찾기
scroll_area = driver.find_element_by_xpath('//*[@id="yDmH0d"]/c-wiz/div/div[2]/div[2]/div[4]/div/div/div[2]/div/div[1]')

#ㅎ해당 영역까지 일단 전체 화면을 스크롤
action = webdriver.ActionChains(driver)
action.move_to_element(scroll_area).perform()

# 해당 영역 내부를 제일 하단까지 스크롤
driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight",scroll_area)

covid_list=soup.select("tr.sgXwHf.wdLSAe")

# 필요한 정보를 출력하는 함수
def printInfo(index, loc, tds):
    total = tds[0].text
    new = tds[1].text
    confirmed_per_mil = tds[3].text
    death = tds[4].text
    
    print(f"[{index:>3}] 위치: {loc} 총 확진자 수: {total} | 신규확진자(1일*): {new} \
    | 백만 명당 확진자 수: {confirmed_per_mil} | 사망: {death} ")
    
# 전체 리스트에서 각 정보를 출력
def printInfoAll(covid_list):
    for index, covid_info in enumerate(covid_list):
        loc = covid_info.find('div').text
        tds = covid_info.find_all('td')
        printInfo(index, loc, tds)
driver.close()
```

```python
printInfoAll(covid_list)
```
